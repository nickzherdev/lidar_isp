```python
#!/usr/bin/env python3
import matplotlib.pyplot as plt
import numpy as np
from scipy.linalg import cholesky
```

```python
def get_ellipse(a, b, n):
    '''Return `n` points of an ellipse defined by `a` and `b`'''
    phi = np.linspace(0, 2*np.pi, n)
    r = a*b/np.sqrt((b*np.cos(phi))**2 + (a*np.sin(phi))**2)
    x = r*np.cos(phi)
    y = r*np.sin(phi)
    return np.vstack([x, y])

def get_noisy_ellipse(sigma=0.01, a=0.5, b=1, n=100):
    '''Return `n` points of an ellipse defined by `a` and `b` with added
    normal noise with standard deviation `sigma`'''
    xy = get_ellipse(a, b, n)
    return xy + np.random.normal(scale=sigma, size=xy.shape)

def get_rt(x):
    '''Get rototranslation from vector of deltas `[dx, dy, d_theta]`'''
    s, c = np.sin(x[2]), np.cos(x[2])
    r = np.array([[c, s], [-s, c]])
    t = np.array(x[:2])
    return r, t

def transform(points, rt):
    '''Transform `points` using given rototranslation `rt`'''
    r, t = rt
    return r.dot(points) + t[:, np.newaxis]

def calculate_distance(coords1, coords2):
    distance_map_frame = coords2[:2] - coords1[:2]
    return distance_map_frame

def corresp_icp_step(points, ref_points):
    '''Perform ICP iteration on corresponding points. Length of both lists
    must be equal.
    
    Input: 
    Output: rotation vector R, thanslation vector T

    Calculate centre of masses for each point cloud [cm1_x, cm1_y], [cm2_x, cm2_y]
    Subtract from each point it's cloud CoM
    
    Find roto-translation
    r = [[cos(x), sin(x)], 
         [-sin(x), cos(x)]]
    matrix = sum(p_i*r.T)
    
    np.linalg.svd or Cholesky on matrix -- L which is the rotation matrix?
    consult ps1 perceprion
    
    trans = mu_rp - R * mu_p # or vice versa
    
    '''
    
    # change shape of points to [x1, y1], [x2, y2]
    points = points.T
    ref_points = ref_points.T
    
    assert points.shape[1] == 2
    assert ref_points.shape[1] == 2
    
    # calculate centre of mass
    mean_pc1 = np.mean(points, axis=0)
    mean_pc2 = np.mean(ref_points, axis=0)
    
    # subtract from each point in PC it's centre of mass
    points_ = points - mean_pc1
    ref_points_ = ref_points - mean_pc2
    
    # W should be 2x2 matrix
    W = np.dot(points_.T, ref_points_)
    
#     L = cholesky(W.T, lower=True) # rotation matrix
#     print("cholesky L")
#     print(L * L.T)
    
    # W.T !!!! 
    U, S, V = np.linalg.svd(W.T)
    R = np.dot(U, V)
#     print()
#     print("SVD R")
#     print(R)
    t = mean_pc2 - np.dot(R, mean_pc1)
    
    return R, t

def find_corresp(points, ref_points, thresh=20):
    '''Find correcpondences between two point clouds using given threshold.
    Must return two point clouds with equal length.
    If dist_norm lower than threshold, append to crsp_rp

    TODO
    k-d tree reduces complexity to n*log(n)

    Take point from the first array and find point in second array so the dist is minimum
    
    Input: two nparray of points 
    [[x1, x2, x3], 
     [y1, y2, y3]]
    
    Output: two numpy arrays shape = 2x80
            first point in first array corresponds to first point in second array, r is ref
    '''
    
    points = points.T
    ref_points = ref_points.T
    
    assert points.shape[1] == 2
    assert ref_points.shape[1] == 2
    
    crsp_rp = []
    for i, p in enumerate(points):
        distances = []
        for rp in ref_points:
            dist_map_frame = calculate_distance(p, rp)
            dist_norm = np.linalg.norm(dist_map_frame)
            distances.append(dist_norm)
        min_index = np.argmin(distances)
        min_dist = distances[min_index]
        if min_dist < thresh:
            crsp_rp.append(ref_points[min_index])
    crsp_rp = np.array(crsp_rp)

    # TODO
    # here maybe a trick: if point in points doesn't have correspondaces, we shouldn't return it!
    return points.T, crsp_rp.T

def icp_step(points, ref_points, guess, plot=False):
    points = transform(points, guess)
    crsp_p, crsp_rp = find_corresp(points, ref_points)

    if plot:
        plt.clf()
        plt.plot(ref_points[0, :], ref_points[1, :], '.', label="reference scan")
        plt.plot(points[0, :], points[1, :], '.', label="original scan")

        for (cp, crp) in zip(crsp_p.T, crsp_rp.T):
            plt.plot([cp[0], crp[0]], [cp[1], crp[1]], 'black')
        plt.axis('equal')
        plt.legend()
        plt.grid()
    
    r, t = corresp_icp_step(crsp_p, crsp_rp)
    gr, gt = guess
    return r.dot(gr), r.dot(gt) + t

if __name__ == "__main__":
    rp = get_noisy_ellipse()
    p = transform(get_noisy_ellipse(n=80), get_rt([-0.2, 0.3, 0.5]))

    plt.ion()

    rt = np.identity(2, dtype=np.float64), np.zeros(2, dtype=np.float64)

    for _ in range(15):
        plt.figure()
        rt = icp_step(p, rp, rt, plot=True)
```
